```python
import cv2
import time

def take_photo(filename='photo.jpg', quality=0.8, capture_duration=10):
    # Inisialisasi kamera
    cap = cv2.VideoCapture(0)
    
    start_time = time.time()
    while (time.time() - start_time) < capture_duration:
        # Baca frame dari kamera
        ret, frame = cap.read()
        
        # Tampilkan frame dalam jendela video
        cv2.imshow('Camera', frame)
        
        # Tekan 'q' untuk keluar
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    
    # Simpan frame terakhir sebagai gambar
    cv2.imwrite(filename, frame, [int(cv2.IMWRITE_JPEG_QUALITY), int(quality * 100)])
    
    # Bebaskan sumber daya kamera
    cap.release()
    cv2.destroyAllWindows()
    
    return filename

# Ambil foto dengan nama 'photo.jpg' dengan durasi pengambilan gambar 10 detik
take_photo('photo.jpg', quality=0.8, capture_duration=5)
```
